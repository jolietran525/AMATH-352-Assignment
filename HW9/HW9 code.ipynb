# **Displacement model**

```python
# K = 2
# Construct matrix X1
X1 = np.vander(data['X_train'][:,0], 3)

# Condition number
K_X12 = np.linalg.cond(X1)
rankX12 = np.linalg.matrix_rank(X1)
sizeX12 = X1.shape

# Compute the alpha coefficients 
Q, R = np.linalg.qr(X1)
alpha_X12 = np.linalg.solve(R, np.dot(Q.T, data['Y_train']))

# Compute predicted value on test dataset
X1_test = np.vander(data['X_test'][:, 0], 3)
y1_test = X1_test.dot(alpha_X12)
RE_12 = ( 1/(np.linalg.norm(data['Y_test'])**2) )*np.linalg.norm(y1_test - data['Y_test'])**2
```
